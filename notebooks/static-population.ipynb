```python
import pandas as pd
import numpy as np
```

```python
population = pd.read_csv("https://raw.githubusercontent.com/MoH-Malaysia/covid19-public/main/static/population.csv")
print(population.columns)
population
```

```python
population['state'].replace({"W.P. ":""}, regex=True, inplace=True)
population.loc[population['state']=='Pulau Pinang', 'state'] = 'Penang'
print(population['state'].unique())
population.drop(columns='idxs', axis=1, inplace=True)
population.drop(0, inplace=True)
population
```

```python
population.to_csv("../data/static-population.csv", index=False)
```

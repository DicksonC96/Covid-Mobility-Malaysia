```python
import pandas as pd
```

```python
waze = pd.read_csv("https://raw.githubusercontent.com/ActiveConclusion/COVID19_mobility/master/waze_reports/waze_mobility.csv")
print(waze.columns)
waze.head()
```

```python
print("Malaysia" in waze['country'].unique())
malaysia = waze.loc[(waze['country']=='Malaysia')]
malaysia.describe(include="all")
```

```python
print(malaysia['city'].unique())
wazedata = malaysia[['date', 'city', 'driving_waze']]
wazedata.describe(include="all")
wazedata.to_csv("waze-mobility-data-malaysia.csv", index=False)
```
